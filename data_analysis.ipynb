```python
import pandas as pd
import matplotlib.pyplot as plt
```

Tick Data Analysis 

```python
# Numerical Anaylsis of dataset  
df = pd.read_csv('datasets/TICK_project_data.csv')

# ================================================================================
# Status
status_data = {
    "Augment": 0,
    "Discard": 0,
    "Keep": 0,
}
for index, row in df.iterrows():
        if row["Status"] == "Augment":
            status_data["Augment"] += 1
        if row["Status"] == "Discard":
            status_data["Discard"] += 1
        if row["Status"] == "Keep":
            status_data["Keep"] += 1
print(status_data)

# ================================================================================
# Pixel Size 
pixel_size_counter = 0
for index, row in df.iterrows():
        if row["Pixel size"] != None:
           row["Pixel size"] += 1    
print(pixel_size_counter)

# ================================================================================
# Sex
sex_data = {
    "unk": 0,
    "f": 0,
    "m": 0,
}
for index, row in df.iterrows():
        if row["Sex"] == "unk":
            sex_data["unk"] += 1
        if row["Sex"] == "f":
            sex_data["f"] += 1
        if row["Sex"] == "m":
            sex_data["m"] += 1
print(sex_data)

# ================================================================================
# Lifestage
lifestage_data = {
    "unk": 0,
    "a": 0,
    "n": 0,
}
for index, row in df.iterrows():
        if row["Lifestage"] == "unk":
            lifestage_data["unk"] += 1
        if row["Lifestages"] == "a":
            lifestage_data["a"] += 1
        if row["Lifestage"] == "n":
            lifestage_data["n"] += 1
print(lifestage_data)

# ================================================================================
# Feedstage
feedstage_data = {
    "unk": 0,
    "fed": 0,
    "unfed": 0,
}
for index, row in df.iterrows():
        if row["Feed-Stage"] == "unk":
            feedstage_data["unk"] += 1
        if row["Feed-Stage"] == "fed":
            feedstage_data["fed"] += 1
        if row["Feed-Stage"] == "unfed":
            feedstage_data["unfed"] += 1
print(feedstage_data)

# ================================================================================
# Species
species_data = {
    "Amblyomma_americanum": 0,
    "Dermacentor_variabilis": 0,
    "Ixodes_scapularis": 0,
}
for index, row in df.iterrows():
        if row["Species"] == "Amblyomma_americanum":
            species_data["Amblyomma_americanum"] += 1
        if row["Species"] == "Dermacentor_variabilis":
            species_data["Dermacentor_variabilis"] += 1
        if row["Species"] == "Ixodes_scapularis":
            species_data["Ixodes_scapularis"] += 1
print(species_data)
```
